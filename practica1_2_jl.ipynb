```julia
using Random
using DelimitedFiles
```

```julia
# Insertion Sort (ordenamiento por inserción)
function insertion_sort!(A)

    n = length(A)  # tamaño del arreglo

    # Bucle principal
    for j in 2:n
        key = A[j]       # elemento a insertar
        i = j - 1

        # Desplazar elementos mayores que la clave
        while i >= 1 && A[i] > key
            A[i + 1] = A[i]
            i -= 1
        end

        # Insertar en la posición correcta
        A[i + 1] = key
    end

    return A
end
```

```julia
# Generar un arreglo pequeño
data = rand(10)

println("Datos originales:")
println(data)

# Ordenar
insertion_sort!(data)

println("\nDatos ordenados:")
println(data)

println("\n¿Está ordenado correctamente?: ", issorted(data))
```

```julia
# Tamaños de prueba
valores_N = [100, 1000, 5000, 10000]

tiempos = []

println("Pruebas de rendimiento Insertion Sort")

for N in valores_N

    # Generar datos aleatorios
    data = rand(N)

    # Medir tiempo
    tiempo = @elapsed insertion_sort!(data)

    push!(tiempos, tiempo)

    println("N = ", N, "   Tiempo = ", tiempo, " segundos")
end
```
